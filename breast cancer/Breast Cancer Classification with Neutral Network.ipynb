# Data collection & Processing 

```python
# Loading data from sklearn 
breast_cancer_dataset = sklearn.datasets.load_breast_cancer()
```

```python
# loading the data to a data frame 
data_frame = pd.DataFrame(breast_cancer_dataset.data, columns = breast_cancer_dataset.feature_names)
```

```python
# Adding the 'target' column to the data fame
data_frame['label'] = breast_cancer_dataset.target
```

```python
# Seperating the features and target
X = data_frame.drop(columns='label', axis=1)
Y = data_frame['label']
```

```python
print(X)
```

```python
print(Y)
```

```python
# Spilting the data into training and testing 
X_train, X_test, Y_train, Y_test = train_test_split (X,Y, test_size=0.2, random_state=2)
print(X.shape, X_train.shape, X_test.shape)
```

```python
#Standardize the data
from sklearn.preprocessing import StandardScaler
scaler = StandardScaler()
X_train_std = scaler.fit_transform(X_train) #Fit scaler to your train data alone and not your test data
X_test_std = scaler.transform(X_test)
print(X_train_std)
```

Building the Neutral Network

```python
#importing tensorflow and Keras
import tensorflow as tf
tf.random.set_seed(3)
from tensorflow import keras
```

```python
# setting up the layers of neutral network 
model = keras.Sequential ([keras.layers.Flatten(input_shape=(30,)),  #Input layer
                           keras.layers.Dense(20, activation='relu'), #middle layer
                           keras.layers.Dense(2, activation='sigmoid') #output layer. The number of neurons equals to the number of output layer
])
```

```python
# compiling the neutral network 
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy', 
              metrics=['accuracy'])
```

```python
# Training the neutral network 
# Adding X_train_std to have a higher accuracy score to your model
history = model.fit(X_train_std, Y_train, validation_split=0.1, epochs=10)
```

```python
# Visualizing Accuracy and loss overtime
plt.plot(history.history['accuracy'])
plt.plot(history.history['val_accuracy'])

plt.title('model accuracy')
plt.ylabel('accuracy')
plt.xlabel('epoch')
plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
plt.plot(history.history['loss'])
plt.plot(history.history['val_loss'])

plt.title('model accuracy')
plt.ylabel('loss')
plt.xlabel('epoch')

plt.legend(['training data', 'validation data'], loc = 'lower right')
```

```python
#Accuracy of the model on test data
loss, accuracy = model.evaluate(X_test_std, Y_test)
print(accuracy)
```

```python
#Model.predict gives the prediction probability of each class of the data point. 

#converting the prediction probability to class label 
Y_pred_labels = [np.argmax(i) for i in Y_pred]
print(Y_pred_labels)
```
